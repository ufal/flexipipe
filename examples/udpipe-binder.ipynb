# Flexipipe UDPipe (Binder-ready)

This notebook runs a minimal UDPipe REST example via Flexipipe. It is designed to work on mybinder.org using the `binder/requirements.txt` in this repo.

What it does:
- Runs UDPipe tagging/parsing over a short text via Flexipipe
- Prints token, lemma, UPOS, XPOS, and dependencies

No local models are downloaded; UDPipe REST is used.



```python
from flexipipe.backends import udpipe

text = "This is a small Binder-based UDPipe test."
model = "english-ewt-ud-2.17-251125"  # curated model name

# Run UDPipe REST via Flexipipe to get a Document
# (No local model download needed; uses the public REST endpoint)
doc = udpipe.tag(text, model=model)

print("Sentences:", len(doc.sentences))
for sent in doc.sentences:
    print("---")
    for tok in sent.tokens:
        print(tok.text, tok.lemma, tok.upos, tok.xpos, tok.head, tok.deprel)

```
